```python
from google.colab import files
import os

uploaded = files.upload()

# Get uploaded file path
audio_path = next(iter(uploaded))
print(f"Uploaded file: {audio_path}")
```

```python
import whisper

# Load the model
model = whisper.load_model("base")

# Load and preprocess audio
audio = whisper.load_audio(audio_path)
audio = whisper.pad_or_trim(audio)

# Get spectrogram
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# Detect language
_, probs = model.detect_language(mel)
language_code = max(probs, key=probs.get)

# Map language codes to full names
language_names = {
    "en": "English",
    "zh": "Chinese",
    "de": "German",
    "es": "Spanish",
    "ru": "Russian",
    "ko": "Korean",
    "fr": "French",
    "ja": "Japanese",
    "pt": "Portuguese",
    "tr": "Turkish",
    "pl": "Polish",
    "ca": "Catalan",
    "nl": "Dutch",
    "ar": "Arabic",
    "sv": "Swedish",
    "it": "Italian",
    "id": "Indonesian",
    "hi": "Hindi",
    "fi": "Finnish",
    "vi": "Vietnamese",
    "he": "Hebrew",
    "uk": "Ukrainian",
    "el": "Greek",
    "ms": "Malay",
    "cs": "Czech",
    "ro": "Romanian",
    "da": "Danish",
    "hu": "Hungarian",
    "ta": "Tamil",
    "no": "Norwegian",
    "th": "Thai",
    "ur": "Urdu",
    "hr": "Croatian",
    "bg": "Bulgarian",
    "lt": "Lithuanian",
    "la": "Latin",
    "sl": "Slovenian",
    "kn": "Kannada",
    "sk": "Slovak",
    "te": "Telugu",
    "fa": "Persian",
    "lv": "Latvian",
    "bn": "Bengali",
    "sr": "Serbian",
    "az": "Azerbaijani",
    "sq": "Albanian",
    "sw": "Swahili",
    "gl": "Galician",
    "mr": "Marathi",
    "pa": "Punjabi",
    "gu": "Gujarati",
    "ka": "Georgian",
    "am": "Amharic",
    "hy": "Armenian",
    "si": "Sinhala",
    "km": "Khmer",
    "mk": "Macedonian",
    "ne": "Nepali",
    "mn": "Mongolian",
    "bs": "Bosnian",
    "my": "Burmese",
    "zh-cn": "Chinese (Simplified)",
    "zh-tw": "Chinese (Traditional)",
}

# Show full name
language_full_name = language_names.get(language_code, language_code)
print(f"\n🌍 Detected Language: {language_full_name} ({language_code})")
```
