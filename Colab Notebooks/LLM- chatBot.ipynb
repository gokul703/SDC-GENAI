```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

# Load FLAN-T5
pipe = pipeline("text2text-generation", model="google/flan-t5-large", max_length=256, do_sample=True)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Prompt for conversation
prompt = PromptTemplate(
    input_variables=["history", "user_input"],
    template="""
You are a friendly and helpful AI assistant. Continue the conversation naturally.

Conversation so far:
{history}
User: {user_input}
AI:"""
)

# Chain for LangChain
chat_chain = LLMChain(llm=llm, prompt=prompt)

# Chat history tracker
chat_history = []

def chatbot(user_input, history=[]):
    global chat_history
    # Combine previous messages
    history_text = "\n".join(chat_history[-6:])
    result = chat_chain.invoke({"history": history_text, "user_input": user_input})
    response = result['text']
    # Update history
    chat_history.append(f"User: {user_input}")
    chat_history.append(f"AI: {response}")
    return response
```

```python
import gradio as gr

gr.ChatInterface(fn=chatbot, title="ðŸ¤– AI Chatbot (Local LLM, No API Key)").launch(share=True)
```
