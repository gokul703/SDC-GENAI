```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline
```

```python
source_lang = input("Enter source language code (e.g., en, fr, hi): ").strip()
target_lang = input("Enter target language code (e.g., fr, en, de): ").strip()
text_to_translate = input("Enter the text you want to translate: ").strip()

model_name = f"Helsinki-NLP/opus-mt-{source_lang}-{target_lang}"

```

```python

try:
    translator_pipeline = pipeline("translation", model=model_name)
    llm = HuggingFacePipeline(pipeline=translator_pipeline)
    result = llm.invoke(text_to_translate)
    print("\n🔁 Translated Text:", result)

except Exception as e:
    print("❌ Translation failed. Possible reasons:")
    print("- Invalid language pair")
    print("- Model not available for selected languages")
    print("- Typo in language codes")
    print("Error:", e)

```
