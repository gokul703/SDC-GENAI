```python
# Import the necessary library
from googletrans import Translator

# Initialize the Translator object
translator = Translator()

# Take user input for the text in English
english_text = input("Enter the text in English: ")

# Translate the text to Tamil
translated_text = translator.translate(english_text, src='en', dest='ta')

# Display the translated text
print(f"Translated Text in Tamil: {translated_text.text}")
```
