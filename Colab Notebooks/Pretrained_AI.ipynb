```python
from transformers import pipeline

# Load a free, pretrained text summarization model
summarizer = pipeline("summarization", model="t5-small")

text = """Artificial intelligence (AI) is intelligence demonstrated by machines,
as opposed to human intelligence. AI includes applications such as natural language
processing, machine learning, and robotics. AI is used in various industries."""

# Generate Summary
summary = summarizer(text, max_length=50, min_length=10, do_sample=False)

print("Summary:", summary[0]['summary_text'])
```
